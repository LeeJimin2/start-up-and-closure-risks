```python
# 상권코드별로 그룹화하고 당월 매출 금액을 합산
grouped_data = data1.groupby('상권_코드')['당월_매출_금액'].sum().reset_index()
print(grouped_data)
```

```python
# 상권코드별로 그룹화하고 당월 매출 건수를 합산
grouped_data1 = data1.groupby('상권_코드')['당월_매출_건수'].sum().reset_index()
print(grouped_data1)
```

```python
df1 = pd.DataFrame(grouped_data)
df2 = pd.DataFrame(grouped_data1)
```

```python
# 두 데이터프레임을 상권_코드로 병합
merged_df = pd.merge(df1, df2, on='상권_코드')

# 결과 출력
print(merged_df)
```

```python
df1 = pd.DataFrame(merged_df)
```

```python
# 당월_매출_평균 열 추가
df1['당월_매출_평균'] = df1['당월_매출_금액'] / df1['당월_매출_건수']

# 결과 출력
print(df1)
```

```python
# 상권코드별로 그룹화하고 주중 매출 금액를 합산
grouped_data4 = data1.groupby('상권_코드')['주중_매출_금액'].sum().reset_index()
print(grouped_data4)
```

```python
# 상권코드별로 그룹화하고 주중 매출 건수를 합산
grouped_data3 = data1.groupby('상권_코드')['주중_매출_건수'].sum().reset_index()
print(grouped_data3)
```

```python
df3 = pd.DataFrame(grouped_data3)
df4 = pd.DataFrame(grouped_data4)

# 두 데이터프레임을 상권_코드로 병합
merged_df3 = pd.merge(df3, df4, on='상권_코드')

# 결과 출력
print(merged_df3)
```

```python
df3 = pd.DataFrame(merged_df3)
```

```python
# 당월_매출_평균 열 추가
df3['주중_매출_평균'] = df3['주중_매출_금액'] / df3['주중_매출_건수']

# 결과 출력
print(df3)
```

```python
# 두 데이터프레임을 상권_코드로 병합
merged_df1 = pd.merge(df3, df1, on='상권_코드')
```
